```python
import pandas as pd
import requests
import matplotlib.pyplot as plt
import seaborn as sns
from io import StringIO

# Correct PubChem API URL
url = "https://pubchem.ncbi.nlm.nih.gov/rest/pug/compound/cid/2244,1983,702,6322,5957/property/MolecularWeight,XLogP,InChIKey/CSV"

# Load data
response = requests.get(url)
csv_data = StringIO(response.text)
df = pd.read_csv(csv_data)

# Show dataset
print(df.columns)
print(df.head())

# Plot Molecular Weight vs XLogP
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='MolecularWeight', y='XLogP')
plt.title('Molecular Weight vs LogP')
plt.xlabel('Molecular Weight')
plt.ylabel('LogP (XLogP)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(data=df, x='CID', y='XLogP')
plt.title('LogP (XLogP) by Compound')
plt.xlabel('Compound CID')
plt.ylabel('LogP (XLogP)')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

### Bonus: How LogP and Molecular Weight Help in Solvent Selection

LogP (specifically XLogP) reflects how hydrophobic a molecule is. A higher LogP means the molecule prefers oily (non-polar) environments, while a lower LogP indicates it is more water-soluble. In drug development, knowing LogP helps in selecting the right solvent to dissolve and deliver the drug effectively. Molecular weight also plays a role in absorption and formulation decisions. Analyzing these trends helps optimize drug design and formulation strategies.
